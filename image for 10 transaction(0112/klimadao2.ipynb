```python

import pandas as pd
from collections import Counter

# 讀取 CSV 文件
data = pd.read_csv("/content/klima_transfers_desc_1_10000.csv")

# 用於存儲從-from 地址到-to 地址的交易次數
transaction_counter = Counter()

# 遍歷數據中的每一行，並更新交易次數
for index, row in data.iterrows():
    from_address = row['from']
    to_address = row['to']
    # 使用 tuple ('from_address', 'to_address') 作為鍵來計數
    transaction_counter[(from_address, to_address)] += 1

# 顯示交易次數
for transaction, count in transaction_counter.items():
    from_address, to_address = transaction
    print(f"交易從 {from_address} 到 {to_address} 出現了 {count} 次")
```
